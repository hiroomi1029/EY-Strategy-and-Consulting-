```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import os

# トモズウェブページのURLを指定
url = "https://shop.tomods.jp/all"

# ウェブページの内容を取得
response = requests.get(url)

# HTTPステータスコードが正常ならばデータを解析
if response.status_code == 200:
    # BeautifulSoupを使用してHTMLを解析
    soup = BeautifulSoup(response.text, "html.parser")
    
    # button要素の中でテキストに "トモズ" を含み、かつ "薬局トモズ" ではないものを全て検索
    target_buttons = soup.find_all("button", text=lambda text: text and "トモズ" in text and "薬局トモズ" not in text)

    # 各ボタン内の住所を取得
    if target_buttons:
        for target_button in target_buttons:
            shop_name = target_button.text.replace("店舗名:", "").strip()  # "店舗名:" を削除し、余分な空白を削除
            shop_info = target_button.find_next("div").text  # 次のdiv要素（住所とその他情報を含む）を取得
            address = shop_info.split(" - ")[-1].strip()  # " - " で分割し、最後の要素を住所として取得
            
            print(f"{shop_name}: {address}")
    else:
        print("該当する店舗が見つかりませんでした。")

else:
    print("ウェブページにアクセスできませんでした。")

```

```python
# ウェブページのURLを指定
url = "https://shop.tomods.jp/all"

# ウェブページの内容を取得
response = requests.get(url)

# HTTPステータスコードが正常ならばデータを解析
if response.status_code == 200:
    # BeautifulSoupを使用してHTMLを解析
    soup = BeautifulSoup(response.text, "html.parser")
    
    # データを格納するリストを初期化
    shop_data = []

    # button要素の中でテキストに "トモズ" を含み、かつ "薬局トモズ" ではないものを全て検索
    target_buttons = soup.find_all("button", text=lambda text: text and "トモズ" in text and "薬局トモズ" not in text)

    # 各ボタン内の店舗情報を取得
    if target_buttons:
        for target_button in target_buttons:
            shop_name = target_button.text.replace("店舗名:", "").strip()  # "店舗名:" を削除し、余分な空白を削除
            shop_info = target_button.find_next("div").text  # 次のdiv要素（住所とその他情報を含む）を取得
            address = shop_info.split(" - ")[-1].strip()  # " - " で分割し、最後の要素を住所として取得
            
            # リストに店舗名と住所を追加
            shop_data.append([shop_name, address])
    
        # DataFrameを作成
        df7 = pd.DataFrame(shop_data, columns=["店舗名", "店舗情報"])
    else:
        print("該当する店舗が見つかりませんでした。")

else:
    print("ウェブページにアクセスできませんでした。")


# df7をCSVファイルに変換
df7.to_csv('tomozu.csv',index= True, encoding='utf_8_sig')

df7
```

```python
url = "https://www.mapion.co.jp/phonebook/M02021Cc01/3.html"


# ウェブページの内容を取得
response = requests.get(url)

# HTTPステータスコードが正常ならばデータを解析
if response.status_code == 200:
    # BeautifulSoupを使用してHTMLを解析
    soup = BeautifulSoup(response.text, "html.parser")
    
    # データを格納するリストを初期化
    shop_data = []

    # 店舗情報を含む要素を取得
    store_items = soup.find_all("table", class_="list-table")

    # 各店舗情報を取得
    for store_item in store_items:
        store_name = store_item.find("a", title=True)
        store_address = store_item.find("td", class_="table-addr").text.strip()
        # 住所が指定の地域に関連しているか確認
        if "埼玉県" in store_address or "東京都" in store_address or "神奈川" in store_address or "千葉" in store_address:
            # データをリストに追加
            shop_data.append({"店舗名": store_name["title"], "住所": store_address})
    
    # DataFrameを作成
    df7 = pd.DataFrame(shop_data)
    
    # 結果を表示
    print(df7)
else:
    print("ウェブページにアクセスできませんでした。")


```
